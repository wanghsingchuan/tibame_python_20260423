```python
class Date:
    # 設定Date(年,月,日)
    def set_data(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day

    # 確認1.是否符合日期格式
    # 沒有處理大小月與2月閏年+1天，處理方式詳: 20260428
    def is_date(self):
        if (31 >= self.day > 0) and (12 >= self.month > 0) and (self.year >= 1911):
            return True
        else:
            return False

    # 確認2.是否閏年
    # 規則: 每4年1閏, 每100年不閏, 每400年閏
    def is_leap_year(self):
        if self.year % 400 == 0:
            return True
        elif self.year % 100 == 0:
            return False
        elif self.year % 4 == 0:
            return True
        else:
            return False


i = 4499
count = 0
while count < 10:

    d = i * i           #數值d = 20241001拆成年月日
    day = d % 100       #日:   01, 20241001除100取餘數
    d_tmp = d // 100    #除100取商, 替換d = 202410, 用來取月
    month = d_tmp % 100 #月:   10, 202410除100取餘數
    year = d_tmp // 100 #年: 2024, 202410除100取商

    # 設定: 取年year 月month 日day
    date = Date()
    date.set_data(year, month, day)

    # 排除閏年 and 確認為日期樣式: 年(1911-), 月(1-12), 日(1-31)
    if not date.is_leap_year() and date.is_date():
        print('完全平方日:', d, '[', year, '年', month, '月', day,'日 ]')
        count += 1
    i = i + 1
```

```python
# 日期確認
def is_date_valid(d):
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100

    if year >= 1:
        if 12 >= month >= 1:
            month_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            days_max = month_days[month-1]
            if is_leap_year(year) and month == 2:
                days_max = days_max + 1
            if days_max >= day >= 1:
                return True
            else:
                return False
        else:
            return False
    else:
        return False

# print(is_date_valid(200010229))
```

```python
# 主程式
start = 4499
result = []
count = 0

while count < 10:
    date = start ** 2
    year = date // 10000
    if is_date_valid(date):

        # 清單專屬功能(直接使用即生效, 不用設定回去的動作): append
        result.append(date)

        count = count + 1
    start = start + 1

# print(result)
```

```python
# 主程式寫成涵式
def get_squared_2(start=4499, max_count=10):
    result = []
    count = 0

    while count < max_count:
        date = start ** 2
        year = date // 10000
        if is_date_valid(date):

            # 清單專屬功能(直接使用即生效, 不用設定回去的動作): append
            result.append(date)

            count = count + 1
        start = start + 1
    return result

s_dates = get_squared_2(max_count=20)
print(s_dates)
print(s_dates[-2])
```

```python
# 把想法1 變成函式
# len(result) 取代 count 的作法

import datetime as dt

def get_squared_1(y=2024, m=10, d=1, max_count=10):
    start = dt.datetime(y, m, d)
    delta = dt.timedelta(days=1)
    result = []

    while len(result) < max_count:
        # 日期 ->字串 ->整數
        # 開根號確認為整數
        # count + 1
        date = int(start.strftime('%Y%m%d'))
        if (date ** 0.5).is_integer():
            result.append(date)
        start = start + delta

    return result

get_squared_1(max_count=5)
```

# 雙層迴圈

```python
# OXOXO
# XOXOX
# OXOXO

for i in range(5):
    line = ''
    if i % 2 == 0:
        for j in range(5):
            if j % 2 == 0:
                line = line + 'O'
            else:
                line = line + 'X'
    else:
        for j in range(5):
            if j % 2 == 0:
                line = line + 'X'
            else:
                line = line + 'O'
    print(line)
```
